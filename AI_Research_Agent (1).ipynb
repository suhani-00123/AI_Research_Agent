**Initialize the Model**

The model is responsible for understanding prompts and generating responses.

```python
model = genai.GenerativeModel("gemini-3.5-flash")
```

**Test the Connection**

Before building the project, verify that the model is responding correctly.

```python
response = model.generate_content("Hello")

print(response.text)
```

**Create a Reusable Function**

Using a function avoids repeating the same code every time a prompt is sent.

```python
def ask_ai(question):

    response = model.generate_content(question)

    return response.text
```

```python
question = input("Ask Your Question: ")

answer = ask_ai(question)

print(answer)
```

```python
from tavily import TavilyClient
```

**Load Tavily API Key**

The Tavily API key is loaded securely from Colab Secrets.

```python
TAVILY_API_KEY = userdata.get("TAVILY_API_KEY")
```

```python
tavily = TavilyClient(api_key=TAVILY_API_KEY)
```

**Create Search Function**

This function searches the web and returns the latest results.

```python
def search_web(query):

    result = tavily.search(
        query=query,
        max_results=3
    )

    return result
```

```python
result = search_web("Latest AI News")

print(result)
```

**Display Search Results**

Extract only the useful information from the search response.

```python
for item in result["results"]:

    print("=" * 60)

    print("Title :", item["title"])

    print("URL :", item["url"])

    print()

    print(item["content"])

    print()
```

The model decides if the question requires a web search.

```python
def should_search(question):

    prompt = f"""
You are an AI Research Assistant.

Decide whether the following question needs a web search.

Reply with only one word:

YES
or
NO

Question:
{question}
"""

    response = ask_ai(prompt)

    return response.strip().upper()
```

```python
question = input("Ask Your Question: ")

decision = should_search(question)

print("Decision :", decision)
```

**Generate Final Answer**

If a web search is required, use the search results to generate a better response.

```python
def research(question):

    decision = should_search(question)

    if "YES" in decision:

        result = search_web(question)

        context = ""

        for item in result["results"]:
            context += item["content"] + "\n"

        prompt = f"""
Question:
{question}

Search Results:
{context}

Using the search results, write a clear and detailed answer.
"""

        return ask_ai(prompt)

    else:

        return ask_ai(question)
```

```python
question = input("Ask Your Question: ")

answer = research(question)

print("\nAnswer:\n")

print(answer)
```

**Create Chat History**

```python
chat_history = []
```

```python
question = input("Ask Your Question: ")

answer = research(question)

chat_history.append({
    "Question": question,
    "Answer": answer
})

print("\nAnswer:\n")
print(answer)
```

**View Chat History**

```python
for chat in chat_history:

    print("=" * 60)

    print("Question:")
    print(chat["Question"])

    print()

    print("Answer:")
    print(chat["Answer"])

    print()
```

```python
with open("chat_history.txt", "w", encoding="utf-8") as file:

    for chat in chat_history:

        file.write(f"Question: {chat['Question']}\n")

        file.write(f"Answer: {chat['Answer']}\n")

        file.write("-" * 60 + "\n")

print("Chat history saved successfully.")
```
